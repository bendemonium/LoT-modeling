```python
from midutils import Element, KComplexity, ElementSet, Associations
import midcf as cf
```

```python
a = Element('A1','A','1')
b = Element('A2','A','2')
c = Element('B1','B','1')
d = Element('B2','B','2')
```

```python
element_set = {a,b,c,d}
S = ElementSet(element_set)
kc = KComplexity()
cf.iterate(S)

element_set = {a,b,c,d}
S = ElementSet(element_set)
result, elapsed_time = cf.palindrome(S)

print(kc.get_k_complexity())
print(kc.get_prim_counts())
```
